```python
import pandas as pd
import re
from tqdm import tqdm
```

```python
df1 = pd.read_csv('./naver_news/Naver_news_20210101_20211231.csv',encoding="UTF-8-sig", index_col=0)
df2 = pd.read_csv('./naver_news/Naver_news_20220101_20220531.csv',encoding="UTF-8-sig", index_col=0)
df3 = pd.read_csv('./naver_news/naver_news_crawling_20220601to20220630.csv',encoding="UTF-8-sig",index_col=0)
```

```python
(df1.head(), df2.head(), df3.head())
```

```python
df = pd.concat([df1,df2,df3],axis=0)
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
def clean_text(inputString):
    text_rmv = re.sub('[+]','증가', inputString)
    text_rmv = re.sub('[-]','감소', text_rmv)
    text_rmv = re.sub('[↑]','상승', text_rmv)
    text_rmv = re.sub('[↓]','하락', text_rmv)
    text_rmv = re.sub('[^一-龥ㄱ-힣a-zA-Z0-9_]', ' ', text_rmv)
    return text_rmv
```

```python
nn = []
```

```python
mask_news = ['권도형','NFT','코인','코린이','디지털자산','가상화폐','암호화폐','공시줍줍','이번 주 IPO','오늘의 주요일정', '오늘의 레이더M 뉴스','오늘의 증시 메모','오늘 증시','오늘의 증시','오늘 시황','오늘의 시황','장외주식','장학재단','[인사]','e공시 눈에 띄네','주식 매매 상위 종목 및 환율','유가증권시장 공시','강영연의 뉴욕오프닝','결정적 한끗','코스닥 주간메모','유가증권 주간메모','머니투데이 오프라인 헤드라인','청년희망적금','투자대회','설명회','오늘의 주요일정','해외 주식형 펀드','장중 기업공시','신용거래융자','일일펀드동향','국내 주식형 펀드','미래에셋투자와연금센터', '[표]', '[부고]','신한디지털포럼','미국 실업수당','신규 선임']
```

```python
mask_news
```

```python
for i in tqdm(range(len(df.index))):
    for j in mask_news:    
        if j in df.iloc[i,1]:
            nn.append(i)
            #print(i)
    df.iloc[i,1] = clean_text(df.iloc[i,1])
```
